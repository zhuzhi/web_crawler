```python
import pandas
import urllib
import re
from bs4 import BeautifulSoup
```

```python
page_limit = 10000
```

```python
url_lsfh_sina = u"http://vip.stock.finance.sina.com.cn/q/go.php/vInvestConsult/kind/lsfh/index.phtml?order=ssrq%7C1&num=PAGELIMIT"
```

```python
url_lsfh_sina = url_lsfh_sina.replace(u"PAGELIMIT", str(page_limit))
```

```python
page_lsfh_sina = urllib.urlopen(url_lsfh_sina)
page_lsfh_sina = page_lsfh_sina.read().decode("GB18030").encode("UTF-8")
```

```python
page_lsfh_sina = BeautifulSoup(page_lsfh_sina, "lxml", from_encoding="UTF-8")
```

```python
table_lsfh_sina = page_lsfh_sina.find_all("table", {"class":"list_table", "id":"dataTable"})
```

```python
table_lsfh_sina = table_lsfh_sina[0]
```

```python
row_lsfh_sina = table_lsfh_sina.find_all("tr")
```

```python
row_lsfh_sina.pop(0)
```

```python
for i in range(len(row_lsfh_sina)):
    row_data = row_lsfh_sina[i]
    row_data = row_data.find_all('td')
    row_data = [row_element.getText() for row_element in row_data]
    row_data.pop()
    row_lsfh_sina[i] = row_data
```

```python
col_names = ["ZQDM", "ZQMC", "SSRQ", "LJGX", "NJGX", "FHCS", "RZZE", "RZCS"]
```

```python
df_lsfh_sina = pandas.DataFrame(row_lsfh_sina, columns=col_names)
```
